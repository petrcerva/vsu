## LSE (Least-Squares Estimation)


### Odhad parametr콢 $\theta$ :

$$ \theta = (\tilde{X^T} \tilde{X})^{-1} \tilde{X^T} y $$  
kde $\tilde{X}$ je roz코칤콏en치 matice dat $X$ (je roz코칤콏en치 o jeden sloupec s hodnotami 1)

#### P콏칤klad matice $\tilde{X}$ pro line치rn칤 regresi:

$$
    X = 
        \begin{bmatrix}
        1 \\
        2 \\
        3\\
        \end{bmatrix} 
    \implies
        \tilde{X} = 
        \begin{bmatrix}
        1 & 1 \\
        1 & 2 \\
        1 & 3\\
        \end{bmatrix} 
$$


#### P콏칤klad matice  $\tilde{X}$ pro polynomi치ln칤 regresi:

* Ka쬯치 vy코코칤 mocnina $X$ je v p콏칤pad캩 PR pova쬺v치na za **dal코칤 nez치vislou prom캩nnou**.

$$
    X = 
        \begin{bmatrix}
        1 & 1 \\
        2 & 4 \\
        3 & 9 \\
        \end{bmatrix} 
    \implies
    \tilde{X} = 
        \begin{bmatrix}
        1 & 1 & 1 \\
        1 & 2 & 4 \\
